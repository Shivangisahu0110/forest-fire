```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv", header = 1)
```

```python
dataset.info()
```

```python
## Missing value
dataset[dataset.isnull().any(axis=1)]
```

The Dataset converted into two sets based on Region from  122th index, we can make a new column based on the region 

1. "Bejaia Region Dataset"
2. "Sidi-Bel Abbes Region Dataset"

```python
dataset.loc[:122,"Region"]=0
```

```python
dataset.loc[122:,"Region"]=1
```

```python
df= dataset
```

```python
df.head()
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[122:]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[122:]
```

```python
df.columns
```

```python
df.head()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df[["day","month","year","Temperature","RH","Ws"]] = df[["day","month","year","Temperature","RH","Ws"]].astype(int)
```

```python
df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']]= df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv("Cleaned_Algerian_forest_fire_dataset.csv",index = False)
```

# EDA

```python
df_copy = df.drop(["day","month","year"],axis = 1)
```

```python
## plot destiny for all feature

plt.style.use("seaborn")
df_copy.hist(bins=50, figsize= (20,15))
plt.show()
```

```python
# Percentage for pie chart
percentage = df_copy["Classes"].value_counts(normalize = True)* 100
```

```python
percentage
```

```python
# Plotting Piechart

Classlabels = ["Fire","Not fire"]
plt.figure(figsize = (12,7))
plt.pie(percentage, labels=Classlabels, autopct="%1.1f%%")
```
